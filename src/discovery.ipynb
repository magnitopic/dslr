## Correlation analysis

```python
numeric_columns = df.select_dtypes(include="float64").columns
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df[numeric_columns].corr(),
            annot=True,
            cmap='coolwarm',
            vmin=-1,
            vmax=1,
            center=0)
plt.title('Mapa de Calor - Correlaciones')
plt.tight_layout()
```
