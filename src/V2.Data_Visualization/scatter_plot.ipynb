# Scatter plot

We verify the perfect relation between `Astronomy` and `Defense Against the Dark Arts`. And then look for other courses with good correlations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
fileSrc = "../data/dataset_train.csv"
df = pd.read_csv(fileSrc, index_col=0)
```

```python
# ScaterPlot requires all rows have value
df = df.dropna()
```

## Perfect correlation corses

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='Astronomy', y='Defense Against the Dark Arts', 
                hue='Hogwarts House', alpha=0.6)

z = np.polyfit(df['Astronomy'], df['Defense Against the Dark Arts'], 1)
p = np.poly1d(z)
plt.plot(df['Astronomy'], p(df['Astronomy']), "r", alpha=0.3)

plt.xlabel('Astronomy')
plt.ylabel('Defense Against the Dark Arts')

plt.tight_layout()

plt.savefig('../output/scatter_plot.png')
plt.show()
plt.close()
```

## Second strongest correlation

```python
# Crear scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='Flying', y='History of Magic', hue='Hogwarts House', alpha=0.6)

# Aadir lnea de tendencia
z = np.polyfit(df['Flying'], df['History of Magic'], 1)
p = np.poly1d(z)
plt.plot(df['Flying'], p(df['Flying']), "r--", alpha=0.8)

plt.xlabel('Flying')
plt.ylabel('History of Magic')

plt.tight_layout()
plt.savefig('../output/scatter_plot_2.png')
plt.show()
plt.close()
```

# What are the two features that are similar?

The two most similar features are `Astronomy` and `Defense Against the Dark Arts` with a perfect r=-1 relation.

The other similar features are `History of Magic` and `Flying` with a correlation of r=-0.9
